```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

PATHDRIVER="###############"




driver=webdriver.Chrome(PATHDRIVER)

driver.get('https://techwithtim.net')
print(driver.title)
```

```python
search=driver.find_element_by_name("s")  #value name
search.send_keys("test") #inserisco parola da mettere nella barra dove si digita
search.send_keys(Keys.RETURN)  #do l'invio
```

```python
#prendo ilrettangolo che contiene tutte le cose che mi servono
main=driver.find_element_by_id("main")
print(main.text)
```

```python
articoli=main.find_elements_by_tag_name("article")

for i in articoli:
    header=i.find_element_by_class_name("entry-title")
    print(header.text)
```

```python
link=driver.find_element_by_link_text("Python Programming") #link cliccabile
link.click()
elemento=driver.find_element_by_link_text("Beginner Python Tutorials").click() #link cliccabile
```
