Tutorial BeautifulSoup

```python
import bs4,requests, webbrowser
```

```python
link="https://www.subito.it/annunci-sardegna/vendita/usato/?q=tavola"
```

```python
response=requests.get(link)  #scarica la pagina
```

```python
response.raise_for_status() # restituisce errore se tutto e andato male
```

```python
soup=bs4.BeautifulSoup(response.text, 'html.parser') #estrae il testo html intero della pagina
```

Control+shift+c dove passa il mouse ti fa vedere tutto nell'ispeziona

```python
divAnnunci=soup.find('div',class_='jsx-4116751698 items visible')
```

```python
a_annunci=divAnnunci.find_all('a') #cosi prendo tutti i tag a
```

```python
linkAnnunci=[]
for i in a_annunci:
    linkAn=str(i.get('href')) #cosi prendo il valore di href cioe il link
    linkAnnunci.append(linkAn)
print(linkAnnunci)
```
